```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Upload image
uploaded = files.upload()
img_path = next(iter(uploaded))
img = cv2.imread(img_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# a. Brightness Enhancement (increase brightness by adding beta)
bright_img = cv2.convertScaleAbs(img, alpha=1, beta=50)  # beta > 0 increases brightness
print("\na. Brightness Enhanced Image:")
cv2_imshow(bright_img)

# b. Contrast Enhancement (increase contrast by scaling with alpha)
contrast_img = cv2.convertScaleAbs(img, alpha=1.5, beta=0)  # alpha > 1 increases contrast
print("\nb. Contrast Enhanced Image:")
cv2_imshow(contrast_img)

# c. Complement of Image (Negative)
complement_img = 255 - img
print("\nc. Complement (Negative) Image:")
cv2_imshow(complement_img)

# d. Binary Contrast Enhancement (simple thresholding)
_, binary_img = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
print("\nd. Binary Contrast Enhancement:")
cv2_imshow(binary_img)

# e. Brightness Slicing (highlight pixels with intensity between 100 and 200)
brightness_sliced = cv2.inRange(gray, 100, 200)
print("\ne. Brightness Slicing:")
cv2_imshow(brightness_sliced)

# f. Low-pass Filtering (Gaussian Blur to smooth image)
low_pass = cv2.GaussianBlur(img, (5, 5), 0)
print("\nf. Low-pass Filtered Image:")
cv2_imshow(low_pass)

# g. High-pass Filtering (sharpening using kernel)
kernel = np.array([
    [-1, -1, -1],
    [-1,  9, -1],
    [-1, -1, -1]
])
high_pass = cv2.filter2D(img, -1, kernel)
print("\ng. High-pass Filtered Image:")
cv2_imshow(high_pass)
```
